```python
import pandas as pd

# Load the CSV file
file_path = 'HAM10000_metadata_with_masks.csv'
metadata = pd.read_csv(file_path)

# Add the 'class' column based on the 'dx' column
metadata['class'] = metadata['dx'].apply(lambda x: 'malignant' if x == 'mel' else 'benign')

# Drop the columns 'dx_type' and 'dataset'
metadata = metadata.drop(columns=['dx_type', 'dataset'])

# Save the updated dataframe to a new CSV file
updated_file_path = 'two_class_final_metadata.csv'
metadata.to_csv(updated_file_path, index=False)

# Display the first few rows of the updated dataframe
metadata.head()
```
