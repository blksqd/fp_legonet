```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '../Data/HAM/HAM10000_metadata.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Basic information about the dataset
print("\nBasic information about the dataset:")
print(data.info())

# Summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(data.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Distribution of the target variable 'dx'
print("\nDistribution of the target variable 'dx':")
print(data['dx'].value_counts())

# Plot the distribution of the target variable 'dx'
plt.figure(figsize=(10, 6))
data['dx'].value_counts().plot(kind='bar')
plt.title('Distribution of Skin Lesion Types')
plt.xlabel('Lesion Type')
plt.ylabel('Count')
plt.show()

# Age distribution
plt.figure(figsize=(10, 6))
plt.hist(data['age'].dropna(), bins=30, edgecolor='k')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of 'sex'
print("\nDistribution of 'sex':")
print(data['sex'].value_counts())

# Plot the distribution of 'sex'
plt.figure(figsize=(6, 4))
data['sex'].value_counts().plot(kind='bar')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Distribution of 'localization'
print("\nDistribution of 'localization':")
print(data['localization'].value_counts())

# Plot the distribution of 'localization'
plt.figure(figsize=(12, 6))
data['localization'].value_counts().plot(kind='bar')
plt.title('Localization of Lesions')
plt.xlabel('Localization')
plt.ylabel('Count')
plt.show()

# Select only numerical columns for the correlation matrix
numerical_columns = data.select_dtypes(include=[np.number])

# Correlation matrix
print("\nCorrelation matrix:")
print(numerical_columns.corr())

# Heatmap of the correlation matrix
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '../Data/HAM/HAM10000_metadata.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Basic information about the dataset
print("\nBasic information about the dataset:")
print(data.info())

# Summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(data.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Apply dark mode style
plt.style.use('dark_background')

# Plotting the age distribution, localization, and diagnosis in a matrix layout
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Plot the distribution of the target variable 'dx' (diagnosis)
axes[0, 0].bar(data['dx'].value_counts().index, data['dx'].value_counts().values, color='c')
axes[0, 0].set_title('Distribution of Skin Lesion Types')
axes[0, 0].set_xlabel('Lesion Type')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xticklabels(data['dx'].value_counts().index, rotation=45, ha='right')

# Age distribution with different colors for sexes
male_ages = data[data['sex'] == 'male']['age'].dropna()
female_ages = data[data['sex'] == 'female']['age'].dropna()
axes[0, 1].hist(male_ages, bins=30, edgecolor='k', alpha=0.7, color='b', label='Male')
axes[0, 1].hist(female_ages, bins=30, edgecolor='k', alpha=0.7, color='r', label='Female')
axes[0, 1].set_title('Age Distribution by Sex')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].legend()

# Distribution of 'localization'
axes[1, 0].bar(data['localization'].value_counts().index, data['localization'].value_counts().values, color='y')
axes[1, 0].set_title('Localization of Lesions')
axes[1, 0].set_xlabel('Localization')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_xticklabels(data['localization'].value_counts().index, rotation=45, ha='right')

# Hide the empty subplot (bottom-right)
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '../Data/ISIC/ISIC_2020_Training_GroundTruth.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Basic information about the dataset
print("\nBasic information about the dataset:")
print(data.info())

# Summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(data.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# List available styles
available_styles = plt.style.available
print("Available styles:", available_styles)

# Apply a selected style (e.g., 'dark_background')
plt.style.use('fivethirtyeight')

# Plotting the age distribution, localization, and diagnosis in a matrix layout
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Plot the distribution of the target variable 'target' (diagnosis)
axes[0, 0].bar(data['target'].value_counts().index, data['target'].value_counts().values, color='c')
axes[0, 0].set_title('Distribution of Diagnosis')
axes[0, 0].set_xlabel('Diagnosis')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xticklabels(data['target'].value_counts().index, rotation=45, ha='right')

# Age distribution with different colors for sexes
male_ages = data[data['sex'] == 'male']['age_approx'].dropna()
female_ages = data[data['sex'] == 'female']['age_approx'].dropna()
axes[0, 1].hist(male_ages, bins=30, edgecolor='k', alpha=0.7, color='b', label='Male')
axes[0, 1].hist(female_ages, bins=30, edgecolor='k', alpha=0.7, color='r', label='Female')
axes[0, 1].set_title('Age Distribution by Sex')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].legend()

# Distribution of 'anatom_site_general_challenge'
axes[1, 0].bar(data['anatom_site_general_challenge'].value_counts().index, data['anatom_site_general_challenge'].value_counts().values, color='y')
axes[1, 0].set_title('Localization of Lesions')
axes[1, 0].set_xlabel('Localization')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_xticklabels(data['anatom_site_general_challenge'].value_counts().index, rotation=45, ha='right')

# Hide the empty subplot (bottom-right)
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
